```python
black_victims = black_victims[black_victims['Year'] <= 1921]

def remove_punctuation(text):
    return re.sub(r'[^\w\s]', ' ', text)

black_victims['Name'] = black_victims['Name'].apply(remove_punctuation)

def remove_double_space(text):
    return re.sub(r'  ', ' ', text)

black_victims['Name'] = black_victims['Name'].apply(remove_double_space)

black_victims = black_victims[black_victims['Name'].str.split().str.len() > 1]

black_victims
```

Above I've subset the data by the year range relevant to VRT (1865 to 1921). I've I removed punctuation from victim names and removed instances where the victim name was only one name since scraping by individual names results in many false positives.

This resulted in 3514 victims. That's going to take like 90 hours of run-time to scrape, if I do it the same way as I did with Seguin/Rigby... Idk about that.

Here's the build_chron_am_search function I used beforehand:

```python
def build_chron_am_search(row):
    base_url = "https://chroniclingamerica.loc.gov/search/pages/results/list/"
    date1 = row['Year']
    date2 = row['Year'] + 1
    phrasetext = row['Name'].replace(' ', '+')

    search_url = (f'{base_url}?date1={date1}&date2={date2}&searchType=advanced&language='
                  f'&proxdistance=5&rows=1000&ortext=&proxtext=&phrasetext={phrasetext}'
                  f'&andtext=&dateFilterType=yearRange&page=1&sort=date')
    
    return search_url

black_victims['search_url'] = black_victims.apply(build_chron_am_search, axis=1)

black_victims
```

```python
black_victims.to_csv('black_subset_tolnay_beck.csv')
```
